```python
import numpy as np
import matplotlib.pyplot as plt
import math
NOS=100 #Number of samples
NOT=10   #Number of training data
lb=0
ub=1
x = np.linspace(lb, ub,num=NOS)
y = np.sin(2*math.pi*x)
```

```python
x1 = np.random.random(NOT);
r1=np.sin(2*math.pi*x1)+(np.random.random(NOT)-0.5)*0.5;

x2 = np.random.random(NOT);
r2=np.sin(2*math.pi*x2)+(np.random.random(NOT)-0.5)*0.5;

x3 = np.random.random(NOT);
r3=np.sin(2*math.pi*x3)+(np.random.random(NOT)-0.5)*0.5;

x4 = np.random.random(NOT);
r4=np.sin(2*math.pi*x4)+(np.random.random(NOT)-0.5)*0.5;

plt.plot(x1, r1, "o",x, y,'g')
plt.axis([lb, ub, -ub, ub])
plt.title("f(x) and sampled data")
plt.savefig("Chp03_biasvariance_a.jpg")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

degree=5

polyreg1=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg1.fit(x1.reshape(NOT,1),r1)
y1=polyreg1.predict(x.reshape(NOS,1));

polyreg2=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg2.fit(x2.reshape(NOT,1),r2)
y2=polyreg2.predict(x.reshape(NOS,1));

polyreg3=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg3.fit(x3.reshape(NOT,1),r3)
y3=polyreg3.predict(x.reshape(NOS,1));

polyreg4=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg4.fit(x4.reshape(NOT,1),r4)
y4=polyreg4.predict(x.reshape(NOS,1));
```

```python
plt.plot(x, y1,'--b', x, y2,'--b', x, y3,'--b',x, y4,'--b')
plt.plot(x, (y1+y2+y3+y4)/4,'-.r')
plt.axis([lb, ub, -ub, ub])
plt.title("order "+str(degree))   
plt.plot(x, y,'g')
plt.savefig("Chp03_biasvariance_d.jpg")
```
