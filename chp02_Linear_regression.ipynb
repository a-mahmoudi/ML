```python
import numpy as np
import matplotlib.pyplot as plt
import math

NOT=10

x = np.arange(0, 10, 0.1);
w1=3
w0=4
y =w1*x+w0 

x1 = np.random.random(NOT)*10;
r1=(w1*x1+w0)+np.random.random(NOT)*4;
```

$$w_0=\overline{r}-w_1\overline(x)$$


$$w_1= \frac{ \sum_t {x^t r^t }-\overline{x} \overline{r}N}{\sum_t {{(x^t)}^2 }-N(\overline{x})^2}$$





```python
x_b=x1.mean()
r_b=r1.mean()
x_sq=np.square(x1)

w1_p=(np.dot(x1,r1)-(x_b*r_b*NOT))/(x_sq.sum()-NOT*x_b*x_b)
w0_p=r_b-w1_p*x_b;

y2 =w1_p*x+w0_p 

plt.plot(x, y,'g')
plt.scatter(x1, r1, marker='o')
plt.plot(x, y2,'r')
```
